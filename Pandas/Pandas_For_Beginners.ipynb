```python
import pandas as pd

# Print Pandas version to ensure proper install
print(pd.__version__)
```

```python
# Define 3 new series objects of random values
x = pd.Series([1,2,3,4])
y = pd.Series([3,4,5,2])
z = pd.Series([1,3,6,3])

# Print example series output
print(x)

# Define 3 column names to store values
cols = ['X', 'Y', 'Z']

# Define a new dataframe with series and column values 
df = pd.DataFrame(list(zip(x,y,z)), columns=cols)

# Use display to nicely print pandas dataframe
display(df)
```

```python
# Read in csv file as Pandas dataframe
df = pd.read_csv('house_prices.csv', index_col='Id')

# print dataframe
print(df)

# observe difference when using display to print dataframe
display(df)
```

```python
# Print first 10 rows of dataframe
display(df.head(10))

# Print last 5(default) rows of dataframe
display(df.tail())
```

```python
# Print shape (dimensions) of dataframe
print(df.shape, '\n')

# Print variable types of columns in dataframe
print(df.dtypes, '\n')

# Print dataframe column names
print(df.columns, '\n')

# Print index values of dataframe
print(df.index, '\n')

# Print values of dataframe as numpy array
print(df.values, '\n')

# Print general information about dataframe including column names, number of null and type of variables
print(df.info())
```

```python
# Print select columns in dataframe
print(df[['BldgType','SaleType']])
```

```python
# Print select rows in dataframe
display(df.loc[[100,200]])
```

```python
# Filter dataframe based on condition where true
temp = df[df['YrSold']>2008]

# Reset the index of dataframe to begin at zero through n
temp.reset_index(drop=True, inplace=True)

# Display new dataframe
display(temp)
```

```python
# Filter dataframe on multiple conditions within one column
temp = df[df['YrSold'].isin([2008,2009])]
display(temp)
```

```python
# Filter dataframe on multiple conditions across multiple columns
temp = df[(df['YrSold']==2008) & (df['SalePrice']>=150000)]

display(temp)
```

```python
# Show mathematic metrics of numeric columns of dataframe
display(df.describe())
```

```python
# Use pandas math operations on column
print(df['SalePrice'].mean())
```

```python
# Get the unique values in column and number of items per unique value
print(df['Street'].value_counts())
```

```python
# Print unique values in dataframe
print(df['YearBuilt'].unique())
```

```python
# Drop columns from dataframe
temp = df.drop(['Alley', 'PoolArea'], axis=1)
display(temp)
```

```python
# Copy dataframe into new variable
temp = df.copy()

# Perform math operations across column values
values = df['LotArea'] / df['SalePrice']
print(values)

# Create new column from list 
temp['Area per SalePrice'] = values
display(temp)
```

```python
# Copy dataframe
temp = df.copy()
display(temp)

# Fill in missing values with preset number
temp.fillna(-10000, inplace=True)
display(temp)
```

```python
# Sort values based on column values 
temp = df.sort_values(['SalePrice', 'YrSold'], ascending=False)
display(temp)
```

```python
# Save new dataframe to csv
temp.to_csv('house_prices_2.csv', index=False)
```
